```python
# deleted unwanted columns

headway_f = headway[['DATE','ROUTE_ABBR','TIME_POINT_ABBR','OPERATOR','TRIP_ID','ROUTE_DIRECTION_NAME','SCHEDULED_TIME', 'ACTUAL_ARRIVAL_TIME',
                     'ACTUAL_DEPARTURE_TIME', 'SCHEDULED_HDWY', 'ACTUAL_HDWY','HDWY_DEV']]

headway_f.head()
```

```python
headway_g = headway_f.sort_values('TIME_POINT_ABBR')
```

```python
# drilled down to just one route, day, and bus stop

headway_h=headway_g[(headway_g['ROUTE_ABBR']==55) & 
        (headway_g['DATE']=='2021-11-01')&
        (headway_g['TIME_POINT_ABBR']=='HHWM')]

headway_h.head(50)
```

```python
# ordered by scheduled time

headway_i = headway_h.sort_values(by='SCHEDULED_TIME', ascending=True)
headway_i.head(50)

# df.sort_values(by='DateTime1',ascending=True)
```

```python

headway_i['DEV_PERCENT'] = headway_i.HDWY_DEV/headway_i.SCHEDULED_HDWY * 100
headway_i
```

```python
# bunched if actually headway is 50% of the scheduled headway. So if the scheduled headway is 10 minutes, then it's bunched if the busses are less than 5 minutes apart. 

# gapped if scheudled headway is 10 minutes and the busses are more than 15 minutes apart if scheduled headway is 10 minutes and the actual headway is 15 minutes 

# created column showing bunching vs gapping
# more than 50% = gapping
# less than -50% = bunching

headway_i['BUNCH/GAP']=''

headway_i.loc[headway_i['DEV_PERCENT'] >50, 'BUNCH/GAP']= 'gap'
headway_i.loc[headway_i['DEV_PERCENT'] <-50, 'BUNCH/GAP']= 'bunch'

headway_i.head(50)
```

```python
# now will try on entire data frame to find trends
```
