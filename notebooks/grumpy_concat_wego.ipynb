```python
corr2['ADHERENCE']
```

```python
avadherence = corr2['ADHERENCE'].abs()
avadherence
```

```python
corr2['Adherence AV'] = corr2['ADHERENCE'].abs()
corr2
```

```python
corr2 = corr2.filter(['Adherence AV', 'SCHEDULED_HDWY', 'ACTUAL_HDWY', 'HDWY_DEV'], axis=1)
```

```python
corr2.rename(columns={'Adherence AV': 'Adherence Absolute Value', 'SCHEDULED_HDWY': 'Scheduled Headway','ACTUAL_HDWY':'Actual Headway','HDWY_DEV':'Headway Deviation'}, inplace=True)
```

```python
mask= np.triu(np.ones_like(corr2.corr()))
cmap=sns.color_palette('RdBu',12)
sns.heatmap(corr2.corr(),mask=mask, vmin=-1, vmax=1, annot=True, cmap=cmap)
plt.title("Correlation Between Adherence & Headway")
plt.savefig("Plotting_Correlation_HeatMap.png",transparent=True,bbox_inches='tight')
```

```python
df=corr2
fig, ax=plt.subplots(figsize=(14,10))
sns.scatterplot(ax=ax,data=df, x = 'Adherence Absolute Value', y= 'Headway Deviation',hue='Headway Deviation',palette="RdBu")
plt.title("Adherence (lateness) VS. Headway")
plt.savefig("Plotting_Correlation_Scatter.jpg",transparent=True,bbox_inches='tight')
plt.show()
```
